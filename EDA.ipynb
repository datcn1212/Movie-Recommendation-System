<h1> EDA

```python
from utils import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies = get_movies()
ratings = get_ratings()
users = get_users()
```

<h2> EDA về người dùng

```python
ax = users["age"].plot(kind="hist", color="slategrey", bins=30, figsize=(10, 10), fontsize=20, rot=0)
ax.set_title("Number users by age", fontsize=24)
ax.set_xlabel("Age", fontsize=20, loc="right")
ax.set_ylabel("Number", fontsize=20, loc="top")
```

```python
ax = ratings.groupby("user_id")["movie_id"].count().plot(kind="hist", bins=30, color="slategrey", figsize=(15, 10), fontsize=20, rot=0)
ax.set_title("Number ratings of each users", fontsize=24)
ax.set_xlabel("Number ratings", fontsize=20, loc="right")
ax.set_ylabel("Frequency", fontsize=20, loc="top")
```

```python
ratings.groupby("user_id")["movie_id"].count().describe()
```

```python
ratings.groupby("user_id")["rating"].mean().describe()
```

```python
ax = ratings.groupby("user_id")["rating"].mean().plot(kind="hist", bins=30, color="slategrey", figsize=(15, 10), fontsize=20, rot=0)
ax.set_title("Average ratings of each users", fontsize=24)
ax.set_xlabel("Average ratings", fontsize=20, loc="right")
ax.set_ylabel("Frequency", fontsize=20, loc="top")
```
